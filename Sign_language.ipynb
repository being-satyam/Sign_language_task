```python


"""
=====================================================================
AFTER RUNTIME RESTART, RUN THIS CELL AND ALL CELLS BELOW
=====================================================================
"""

# =====================================================================
# SECTION 2: IMPORT LIBRARIES
# =====================================================================

print("\n" + "="*70)
print("SECTION 2: IMPORTING LIBRARIES")
print("="*70)

import sys
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import cv2
import time
import warnings
warnings.filterwarnings('ignore')

# Check if in Colab
try:
    from google.colab import files
    from IPython.display import display, Image as IPImage, clear_output
    IN_COLAB = True
    print("✓ Running in Google Colab")
except ImportError:
    IN_COLAB = False
    print("✓ Running in standard Python environment")

print(f"\n✓ Python: {sys.version.split()[0]}")
print(f"✓ NumPy: {np.__version__}")
print(f"✓ TensorFlow: {tf.__version__}")
print(f"✓ OpenCV: {cv2.__version__}")
print(f"✓ Matplotlib: {matplotlib.__version__}")

# Check GPU
try:
    gpu_devices = tf.config.list_physical_devices('GPU')
    gpu_available = len(gpu_devices) > 0
    print(f"✓ GPU Available: {gpu_available}")
    if gpu_available:
        print(f"✓ GPU: {gpu_devices[0].name}")
        for gpu in gpu_devices:
            tf.config.experimental.set_memory_growth(gpu, True)
except:
    gpu_available = False
    print("✓ Running on CPU")

# =====================================================================
# SECTION 3: UPLOAD AND VERIFY DATASET
# =====================================================================

print("\n" + "="*70)
print("SECTION 3: DATASET SETUP")
print("="*70)
print("\nIMPORTANT: Upload your dataset first!")
print("\nOption 1 - Upload ZIP file:")
print("  Run this cell to upload:")
print("  uploaded = files.upload()")
print("  !unzip archive.zip -d /content/ASL_dataset")
print("\nOption 2 - Mount Google Drive:")
print("  from google.colab import drive")
print("  drive.mount('/content/drive')")
print("  DATA_DIR = '/content/drive/MyDrive/ASL_dataset/asl_dataset'")
print("\nExpected structure:")
print("  /content/ASL_dataset/asl_dataset/")
print("    ├── 0/")
print("    ├── 1/")
print("    ├── 2/")
print("    ├── a/")
print("    ├── b/")
print("    └── ... (36 classes total)")

DATA_DIR = "/content/ASL_dataset/asl_dataset"

if not os.path.exists(DATA_DIR):
    print(f"\n❌ ERROR: Dataset not found at {DATA_DIR}")
    print("\n🔧 Quick fix - Run these commands:")
    print("  # Upload your archive.zip file first, then:")
    print("  !unzip /content/archive.zip -d /content/ASL_dataset")
    print("  # OR mount Google Drive:")
    print("  from google.colab import drive")
    print("  drive.mount('/content/drive')")
    raise FileNotFoundError(f"Dataset not found at {DATA_DIR}")

print(f"\n✓ Dataset directory found: {DATA_DIR}")

classes = [d for d in os.listdir(DATA_DIR)
          if os.path.isdir(os.path.join(DATA_DIR, d)) and d != 'asl_dataset']

print(f"✓ Number of classes found: {len(classes)}")
print(f"✓ Classes: {sorted(classes)}")

total_images = 0
for cls in classes:
    cls_path = os.path.join(DATA_DIR, cls)
    try:
        images = [f for f in os.listdir(cls_path)
                 if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))]
        total_images += len(images)
    except:
        pass

print(f"✓ Total images: {total_images}")
if len(classes) > 0:
    print(f"✓ Average per class: {total_images // len(classes)}")

# =====================================================================
# SECTION 4: DATA PREPARATION
# =====================================================================

print("\n" + "="*70)
print("SECTION 4: DATA PREPARATION")
print("="*70)

IMG_SIZE = 300
BATCH_SIZE = 32
VALIDATION_SPLIT = 0.2

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    shear_range=0.1,
    fill_mode='constant',
    cval=255,
    validation_split=VALIDATION_SPLIT
)

print("\nCreating data generators...")
train_generator = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

val_generator = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)

num_classes = len(train_generator.class_indices)
class_names = list(train_generator.class_indices.keys())

print(f"\n✓ Training samples: {train_generator.samples}")
print(f"✓ Validation samples: {val_generator.samples}")
print(f"✓ Classes detected: {num_classes}")
print(f"✓ Class names: {sorted(class_names)}")
print(f"✓ Batch size: {BATCH_SIZE}")
print(f"✓ Steps per epoch: {train_generator.samples // BATCH_SIZE}")

# Visualize samples
print("\nGenerating sample images...")
try:
    fig, axes = plt.subplots(2, 5, figsize=(15, 6))
    sample_batch, sample_labels = next(train_generator)

    for i, ax in enumerate(axes.flat):
        if i < len(sample_batch):
            ax.imshow(sample_batch[i])
            label_idx = np.argmax(sample_labels[i])
            ax.set_title(f"Class: {class_names[label_idx]}")
            ax.axis('off')

    plt.tight_layout()
    plt.savefig('sample_images.png', dpi=150, bbox_inches='tight')
    if IN_COLAB:
        plt.show()
    plt.close()
    print("✓ Sample images saved")
except Exception as e:
    print(f"Warning: Could not generate samples: {e}")

# =====================================================================
# SECTION 5: MODEL ARCHITECTURE
# =====================================================================

print("\n" + "="*70)
print("SECTION 5: MODEL ARCHITECTURE")
print("="*70)

def build_asl_model(img_size=300, num_classes=36):
    """CNN model for ASL recognition"""
    model = keras.Sequential([
        layers.Input(shape=(img_size, img_size, 3)),

        # Block 1
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Block 2
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Block 3
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Block 4
        layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Block 5
        layers.Conv2D(512, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Dense layers
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),

        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),

        layers.Dense(num_classes, activation='softmax')
    ])

    return model

print(f"\n🔧 Building model with {num_classes} classes...")
model = build_asl_model(img_size=300, num_classes=num_classes)

print(f"\n✓ Model built successfully")
print(f"✓ Input: (300, 300, 3)")
print(f"✓ Output: {num_classes} classes")
print(f"✓ Parameters: {model.count_params():,}")

print("\nModel Summary:")
model.summary()

# =====================================================================
# SECTION 6: TRAINING
# =====================================================================

print("\n" + "="*70)
print("SECTION 6: MODEL TRAINING")
print("="*70)

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy', keras.metrics.TopKCategoricalAccuracy(k=5, name='top_5_accuracy')]
)

print("✓ Model compiled")
print("  - Optimizer: Adam (lr=0.01)")
print("  - Loss: Categorical Crossentropy")
print("  - Metrics: Accuracy, Top-5 Accuracy")

MODEL_DIR = "ASL_Model_Paper"
os.makedirs(MODEL_DIR, exist_ok=True)

callbacks = [
    keras.callbacks.ModelCheckpoint(
        filepath=os.path.join(MODEL_DIR, 'best_model.h5'),
        monitor='val_accuracy',
        save_best_only=True,
        mode='max',
        verbose=1
    ),
    keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=15,
        restore_best_weights=True,
        verbose=1
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=7,
        min_lr=1e-7,
        verbose=1
    ),
    keras.callbacks.CSVLogger(
        os.path.join(MODEL_DIR, 'training_log.csv')
    )
]

EPOCHS = 100

print(f"\nStarting training for {EPOCHS} epochs...")
print("This may take several hours...")
print("-" * 70)

start_time = time.time()

history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS,
    callbacks=callbacks,
    verbose=1
)

training_time = time.time() - start_time

print("\n" + "="*70)
print("TRAINING COMPLETED")
print("="*70)
print(f"✓ Training time: {training_time/60:.2f} minutes")
print(f"✓ Total epochs: {len(history.history['accuracy'])}")

final_model_path = os.path.join(MODEL_DIR, 'keras_model.h5')
model.save(final_model_path)
print(f"✓ Model saved: {final_model_path}")

labels_path = os.path.join(MODEL_DIR, 'labels.txt')
with open(labels_path, 'w') as f:
    for label in sorted(class_names):
        f.write(f"{label}\n")
print(f"✓ Labels saved: {labels_path}")

# =====================================================================
# SECTION 7: EVALUATION
# =====================================================================

print("\n" + "="*70)
print("SECTION 7: MODEL EVALUATION")
print("="*70)

results = model.evaluate(val_generator, verbose=1)

print("\n📊 Evaluation Results:")
print(f"  Loss: {results[0]:.4f}")
print(f"  Accuracy: {results[1]:.4f} ({results[1]*100:.2f}%)")
print(f"  Top-5 Accuracy: {results[2]:.4f} ({results[2]*100:.2f}%)")

# =====================================================================
# SECTION 8: VISUALIZATION
# =====================================================================

print("\n" + "="*70)
print("SECTION 8: TRAINING VISUALIZATION")
print("="*70)

try:
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('ASL Recognition Model Training Results',
                 fontsize=16, fontweight='bold')

    # Accuracy
    axes[0, 0].plot(history.history['accuracy'], label='Training', linewidth=2, color='#2E86AB')
    axes[0, 0].plot(history.history['val_accuracy'], label='Validation', linewidth=2, color='#A23B72')
    axes[0, 0].set_xlabel('Epoch', fontsize=12)
    axes[0, 0].set_ylabel('Accuracy', fontsize=12)
    axes[0, 0].set_title('Model Accuracy', fontsize=14, fontweight='bold')
    axes[0, 0].legend(fontsize=11)
    axes[0, 0].grid(True, alpha=0.3)

    # Loss
    axes[0, 1].plot(history.history['loss'], label='Training', linewidth=2, color='#2E86AB')
    axes[0, 1].plot(history.history['val_loss'], label='Validation', linewidth=2, color='#A23B72')
    axes[0, 1].set_xlabel('Epoch', fontsize=12)
    axes[0, 1].set_ylabel('Loss', fontsize=12)
    axes[0, 1].set_title('Model Loss', fontsize=14, fontweight='bold')
    axes[0, 1].legend(fontsize=11)
    axes[0, 1].grid(True, alpha=0.3)

    # Top-5 Accuracy
    axes[1, 0].plot(history.history['top_5_accuracy'], label='Training', linewidth=2, color='#2E86AB')
    axes[1, 0].plot(history.history['val_top_5_accuracy'], label='Validation', linewidth=2, color='#A23B72')
    axes[1, 0].set_xlabel('Epoch', fontsize=12)
    axes[1, 0].set_ylabel('Top-5 Accuracy', fontsize=12)
    axes[1, 0].set_title('Top-5 Accuracy', fontsize=14, fontweight='bold')
    axes[1, 0].legend(fontsize=11)
    axes[1, 0].grid(True, alpha=0.3)

    # Summary
    final_acc = history.history['val_accuracy'][-1]
    best_acc = max(history.history['val_accuracy'])
    summary_text = f"""
📊 TRAINING SUMMARY

Final Validation Accuracy: {final_acc:.4f} ({final_acc*100:.2f}%)
Best Validation Accuracy: {best_acc:.4f} ({best_acc*100:.2f}%)
Total Epochs: {len(history.history['accuracy'])}
Training Time: {training_time/60:.2f} minutes

Dataset: ASL ({num_classes} classes)
Model: CNN (5 Conv Blocks)

Paper: "Real Time Sign Language
Interpretation Leveraging
Computer Vision Technologies"
"""

    axes[1, 1].text(0.5, 0.5, summary_text, ha='center', va='center',
                   fontsize=11, family='monospace',
                   bbox=dict(boxstyle='round', facecolor='#F6F6F6',
                            edgecolor='#2E86AB', linewidth=2))
    axes[1, 1].axis('off')

    plt.tight_layout()
    plot_path = os.path.join(MODEL_DIR, 'training_history.png')
    plt.savefig(plot_path, dpi=300, bbox_inches='tight')
    if IN_COLAB:
        plt.show()
    plt.close()
    print(f"\n✓ Visualization saved: {plot_path}")
except Exception as e:
    print(f"Warning: Could not create visualization: {e}")

# =====================================================================
# SECTION 9: DOWNLOAD FILES
# =====================================================================

print("\n" + "="*70)
print("SECTION 9: DOWNLOAD MODEL FILES")
print("="*70)
print("\nFiles available for download:")
print(f"  1. {MODEL_DIR}/keras_model.h5")
print(f"  2. {MODEL_DIR}/best_model.h5")
print(f"  3. {MODEL_DIR}/labels.txt")
print(f"  4. {MODEL_DIR}/training_log.csv")
print(f"  5. {MODEL_DIR}/training_history.png")

if IN_COLAB:
    print("\nTo download files, run:")
    print("  from google.colab import files")
    print("  files.download('ASL_Model_Paper/keras_model.h5')")

# =====================================================================
# SECTION 10: SUMMARY
# =====================================================================

print("\n" + "="*70)
print("🎉 IMPLEMENTATION COMPLETE!")
print("="*70)
print("\n📝 Summary:")
print(f"  ✓ Dataset: {total_images} images, {num_classes} classes")
print(f"  ✓ Model: CNN with 5 convolutional blocks")
print(f"  ✓ Training: {len(history.history['accuracy'])} epochs")
print(f"  ✓ Best Accuracy: {max(history.history['val_accuracy'])*100:.2f}%")
print(f"  ✓ Training Time: {training_time/60:.2f} minutes")
print("\n📚 Research Paper:")
print("  'Real Time Sign Language Interpretation'")
print("  'Leveraging Computer Vision Technologies'")
print("\n" + "="*70)
print("\n✅ ALL DEPENDENCIES FIXED!")
print("   ml_dtypes 0.5+ | numpy 1.26.4 | TensorFlow 2.17.0")
print("="*70)
```
